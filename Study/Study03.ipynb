## .plot(params)

기본적인 2차원 그래프를 그리는데 사용된다. 

plt.plot(x,y) 식으로 사용되며, plt.plot(param)와 같이 파라미터가 하나로 지정되면 x축은 알아서 0 1 2.. 와 같이 할당된다

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4, 5]
y = [1, 8, 27, 64, 125]

# use .plot() to plot x and y
plt.plot(x,y)

# set labels for x and y
plt.xlabel("input")
plt.ylabel("output")

# use .show() to show the graph
plt.show()
```

또한 그래프의 선 스타일도 지정할 수 있다.

### 점모양

'o' : 원형

'^' : 삼각형

's' : 사각형

'd' : 다이아몬드

'v' : 역삼각형

'*' : 별모양

### 선 스타일

'-' : 실선

'--' : 점선

':' : 대시선

```python
t = np.arange(0, 5, 0.2) # 0~5 0.2 간격 씩 생성

plt.plot(t, t**2, 'r-s', label='square')
plt.plot(t, t**3, color='b', marker='s', label='cubic')

plt.legend() # 범례를 추가함 (label)
plt.show()
```

## Scatter Plot (산점도)

plot와 비슷하게 작동한다. 산점도 형식의 그래프를 그려주는 함수이다.
.scatter(x,y) 형식으로 사용되며, color, marker를 지정할 수 있다.

```python
data = np.random.rand(100, 2) # 100 * 2 무작위 배열 만들기

plt.scatter(data[:,0], data[:,1], color='red', marker='s') # 위에서 만든 값을 x,y 값으로 출력

plt.show()
```

## Bar Plot

막대 그래프 위와 같음 .bar(x,y)

```python
data = np.random.rand(5, 2)

plt.bar(data[:,0], data[:,1], color='b', width=0.01)

plt.show()
```

## Histogram

히스토그램 그리기

```python
x = np.random.randn(1000)

plt.hist(x, bins=100) # bins=100 구간을 지정하는 것

plt.title('Normal Distribution')

plt.show()
```
